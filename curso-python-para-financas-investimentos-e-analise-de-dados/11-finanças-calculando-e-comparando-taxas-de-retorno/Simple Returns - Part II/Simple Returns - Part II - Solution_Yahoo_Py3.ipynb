## Simple Returns - Part II

*Suggested Answers follow (usually there are multiple ways to solve a problem in Python).*

$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

```python
import yfinance as yf

msft_ticker = yf.Ticker("MSFT")
msft_data = msft_ticker.history(start="2000-01-01")

msft_data["Simple Return"] = (
    msft_data["Close"] / msft_data["Close"].shift(1)
) - 1
print(msft_data["Simple Return"])
```

Plot the simple returns on a graph.

```python
import matplotlib.pyplot as plt
```

```python
msft_data["Simple Return"].plot(figsize=(8, 5))
plt.show()
```

Calculate the average daily return.

```python
avg_returns_d = msft_data["Simple Return"].mean()
avg_returns_d
```

Estimate the average annual return.

```python
avg_returns_a = msft_data["Simple Return"].mean() * 250
avg_returns_a
```

Print the percentage version of the result as a float with 2 digits after the decimal point.

```python
print(str(round(avg_returns_a, 4) * 100) + " %")
```

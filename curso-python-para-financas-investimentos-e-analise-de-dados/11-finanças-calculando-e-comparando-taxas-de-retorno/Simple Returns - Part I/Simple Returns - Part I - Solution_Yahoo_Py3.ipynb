Download the data for Microsoft (‘msft_data’) from Yahoo Finance for the period ‘2000-1-1’ until today.

```python
msft_ticker = yf.Ticker("MSFT")
end_date = datetime.now().strftime("%Y-%m-%d")
msft_data = msft_ticker.history(start="2000-01-01", end=end_date)

msft_data
```

### Simple Rate of Return

Calculate the simple returns of ‘msft_data’ for the given timeframe.

$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

```python
msft_data["simple_return"] = (
    msft_data["Close"] / msft_data["Close"].shift(1)
) - 1
print(msft_data["simple_return"])
```
